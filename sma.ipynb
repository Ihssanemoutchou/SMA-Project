```python
import tiktoken
from dotenv.ipython import load_dotenv
load_dotenv(override=True)
import os
print(os.environ['openai_api_key'])
```

```python
tokennizer = tiktoken.encoding_for_model("gpt-4o")
print(tokennizer.name)
```

```python
prompt = "sma77"
```

```python
tokens=tokennizer.encode(prompt)
print(len(tokens))
```

```python
for token in tokens:
    t=tokennizer.decode_single_token_bytes(token).decode("utf-8")
print(t, end="|")
```

```python
def count_tokens(prompt:str ,model:str="gpt-4o"):
    tokennizer = tiktoken.encoding_for_model("gpt-4o")
    tokens=tokennizer.encode(prompt)
    return len(tokens)
```

```python
print(count_tokens(prompt="hello world"))
```
